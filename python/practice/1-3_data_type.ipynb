## 6번
pop과 remove의 차이는 무엇인가?
pop : list의 마지막 요소를 반환하고 삭제, pop(x)와 같이 x index가 파라메터로 전달되는 경우에 해당 index의 요소를 반환하고 그 요소는 삭제함.
remove : remove(x)로 호출되며 x 로 주어진 값에 해당하는 첫번째 요소를 삭제함.
