# Axes Customizing

### fig.tight_layout()

```python
title_list = ['Ax' + str(i) for i in range(4)]
xlabel_list = ['X label ' + str(i) for i in range(4)]
ylabel_list = ['Y label ' + str(i) for i in range(4)]

fig, axes = plt.subplots(2, 2, figsize=(7, 7),
                         facecolor='linen')

for ax_idx, ax in enumerate(axes.flat):
    ax.set_title(title_list[ax_idx], fontsize=20)
    ax.set_xlabel(xlabel_list[ax_idx], fontsize=15)
    ax.set_ylabel(ylabel_list[ax_idx], fontsize=15)
```

```python
title_list = ['Ax' + str(i) for i in range(4)]
xlabel_list = ['X label ' + str(i) for i in range(4)]
ylabel_list = ['Y label ' + str(i) for i in range(4)]

fig, axes = plt.subplots(2, 2, figsize=(7, 7),
                         facecolor='linen')

for ax_idx, ax in enumerate(axes.flat):
    ax.set_title(title_list[ax_idx], fontsize=20)
    ax.set_xlabel(xlabel_list[ax_idx], fontsize=15)
    ax.set_ylabel(ylabel_list[ax_idx], fontsize=15)

fig.tight_layout(pad=1)
```

```python
x = np.arange(-np.pi, np.pi, 0.02)
y1 = np.sin(x)
y2 = np.cos(x)

fig, ax = plt.subplots()

ax.plot(x, y1, label = 'sin') 
ax.plot(x, y2, label = 'cos')

ax.legend(loc=1)

plt.show()
```

```python
x = np.arange(8)
y = x**2

fig, ax = plt.subplots()

ax.plot(x, y, 'ro')

for x_, y_ in zip(x, y):
    ax.text(x_+0.2, y_+0.3, '%d, %d' % (int(x_), int(y_)))
```

annotate

```python
x = np.arange(-1, 3, 0.01)
y = -x**4+4*x**3-4*x**2

fig, ax = plt.subplots()
ax.plot(x, y, lw=2)
ax.annotate('local mininmum', xy=(1, -1), xytext=(-0.5, -3.5),
            arrowprops=dict(facecolor='black'))

ax.set_ylim(-10,2)
plt.show()
```

```python
labels = ['HONEST CANDIDATE', 'Little Women', 'The Closet', 'Jojo Rabbit']
user = [9.2, 9.4, 8.6, 9.16]
critic = [5.4, 8, 5.5, 7.17]
x = np.arange(len(labels))  # the label locations

width = 0.3  # the width of the bars

fig, ax = plt.subplots()
rects1 = ax.bar(x - width/2, user, width, label='user')
rects2 = ax.bar(x + width/2, critic, width, label='critic')

ax.set_ylim(0, 13)
ax.set_ylabel('Rating')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()
```

### Histogram

```python
mu, sigma = 100, 15
x = mu + sigma * np.random.randn(1000)

# the histogram of the data
plt.hist(x, 50, density=True, facecolor='g')

plt.xlabel('IQ')
plt.ylabel('Probability')
plt.title('Histogram of IQ')

plt.text(60, .025, r'$\mu=100,\ \sigma=15$')
plt.xlim(40, 160)
plt.ylim(0, 0.035)
plt.grid(True)
plt.show()
```
