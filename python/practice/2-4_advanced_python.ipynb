### 관리자 상태를 확인하는 코드를 리팩터링

````python
def check_is_admin(username):
    if username != 'admin':
        raise Exception("This user is not allowed to get or put food")

class Store():
    def __init__(self):
        self.storage = ['ice_cream', 'snack']
    
    def get_food(self, username, food):
        check_is_admin(username)
        return self.storage.append(food)

    def put_food(self, username, food):
        check_is_admin(username)
        self.storage.remove(food)

a = Store()
print(a.storage)

a.get_food("admin", "bread")
print(a.storage)

a.get_food("shseo", "beverage")
````

### 데커레이터 사용

````python
def check_is_admin(f):
    def wrapper(*args):
        # print(args)
        if args[1] != 'admin':
            raise Exception("This user is not allowed to get or put food")
        return f(*args)
    return wrapper

class Store():
    def __init__(self):
        self.storage = ['ice_cream', 'snack']
    
    @check_is_admin
    def get_food(self, username, food):
        return self.storage.append(food)

    @check_is_admin
    def put_food(self, username, food):
        self.storage.remove(food)

a = Store()
print(a.storage)

a.get_food("admin", "bread")
print(a.storage)

a.get_food("shseo", "beverage")
````
