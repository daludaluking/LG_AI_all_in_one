### 문제 1
‘students.txt’ 파일을 읽어서 딕셔너리로 학생들의 정보를 저장하라.  
해당 파일은 이름, 국어점수, 영어점수, 수학점수 순으로 되어있으며, 각 컬럼의 구분자는 tab(\t)으로 이루어져있다.  
이름을 key로 하고, 국어점수, 영어점수, 수학점수를 순서대로 담은 리스트를 value로 하는 딕셔너리로 저장하라.

+ read 이용

```python
import os
print(os.getcwd())
file_folder = os.getcwd()
students = {}
```

```python
file_name = 'students.txt'
file_path = os.path.join(file_folder, file_name)
file_path
```

```python
# 정답을 작성해 주세요.
with open(file_path, "r") as fi:
    lines = fi.read().split("\n")
    for line in lines:
        dict_v = line.split()
        students[dict_v[0]] =  [int(v) for v in dict_v[1:]]
    #lines = fi.read().split()
    #for i in range(0, len(lines), 4):
    #    students[lines[i]] = [int(v) for v in lines[i+1:i+4]]
    #while True:
    #    line = fi.readline()
    #    if not line: break
    #    print(line)
    #    kv = line.split()
    #    students[kv[0]] = kv[1:]
#print(students)
```

+ readlines 이용

```python
students2 = {}
```

```python
# 정답을 작성해 주세요.
with open(file_path, "r") as fi:
    while True:
        lines = fi.readlines()
        if not lines: break
        print(lines)
        for line in lines:
            #print(line.split())
            kv = line.split()
            students2[kv[0]] = [int(i) for i in kv[1:]]
```

### 문제 2
저장한 딕셔너리를 토대로 학생들의 평균을 구하라.  
전체 학생의 국어, 영어, 수학 점수의 평균을 구해서 각각 kor, eng, math라는 변수에 저장하고, 출력하라.

```python
students
```

```python
kor = 0
eng = 0
math = 0
```

```python
# 정답을 작성해 주세요.
for v in students2.values():
    print(v)
    kor += int(v[0])
    eng += int(v[1])
    math += int(v[2])

kor /= len(students)
eng /= len(students)
math /= len(students)
```

```python
print('학생들의 평균은 아래와 같습니다.')
print('국어 : ' + str(kor))
print('영어 : ' + str(eng))
print('수학 : ' + str(math))
```

### 문제 3

학생들 개개인의 평균을 구하는 함수를 작성하라.  
이름을 넣으면 국어, 영어, 수학 3과목의 평균을 계산하여 반환하는 함수를 작성하라.

```python
# 정답을 작성해 주세요.
def cal_average(name):
    if name in students2 :
        v = students2[name]
        return sum(v)/len(v)
```
