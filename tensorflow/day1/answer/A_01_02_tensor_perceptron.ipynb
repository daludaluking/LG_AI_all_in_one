```python

import tensorflow as tf

## data 선언
x_data =[[1.],[2.],[3.],[4.]]
y_data =[[1.],[3.],[5.],[7.]]
```

```python
## 평균 0, 분산 1의 파라미터의 정규분포로 부터 값을 가져옴.
# 학습을 통해 업데이트가 되어 변화되는 모델의 파라미터인 w,b를 의미한다.
W=tf.Variable(tf.random.normal((1,1), mean=2, stddev=0.5))
b=tf.Variable(tf.random.normal((1,1), mean=2, stddev=0.5))
print("W : ", W)
print("b : ", b)
```

```python
for j in range(len(x_data)):
    ## data * weight
    WX =tf.matmul([x_data[j]], W)

    ## bias add
    y_hat = tf.add(WX, b)

    ## W와 b로 예측 하기
    print("y_data: , ",y_data[j], "prediction : ", y_hat)
```
