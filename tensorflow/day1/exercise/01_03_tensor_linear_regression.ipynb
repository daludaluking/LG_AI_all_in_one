```python

import tensorflow as tf

## data 선언
x_data =[[1.],[2.],[3.],[4.]]
y_data =[[1.],[3.],[5.],[7.]]

```

```python

## 평균 0, 분산 1의 파라미터의 정규분포로 부터 값을 가져옴.
# 학습을 통해 업데이트가 되어 변화되는 모델의 파라미터인 w,b를 의미한다.
W=tf.Variable(tf.random.normal((1,1),mean=0, stddev=1.0))
b=tf.Variable(tf.random.normal((1,1),mean=0, stddev=1.0))
lr=tf.constant(0.01)
```

```python

for i in range(2000):  ## 에폭
    total_error = 0

    for j in range(len(x_data)): ## 배치 1
        ## data * weight
        WX =tf.matmul([x_data[j]], W)

        ## bias add
        y_hat = tf.add(WX, b)

        ## 정답인 Y와 출력값의 error 계산
        error = tf.subtract(y_data[j], y_hat)
        #error = tf.subtract([y_data[j]], y_hat)

        ## 경사하강법으로 W와 b 업데이트.
        ## 도함수 구하기
        diff_W = tf.matmul(error, [x_data[j]])
        diff_b = error

        ##  업데이트할 만큼 러닝레이트 곱
        ## 중요 multiply로 곱해야함...
        diff_W = tf.multiply(diff_W, lr)
        diff_b = tf.multiply(diff_b, lr)

        ## w, b 업데이트
        W = tf.add(W, diff_W)
        b = tf.add(b, diff_b)
        #######

        ## 토탈 에러.
        total_error = total_error + tf.square(error)

    ## 모든 데이터에 따른 error 값
    print("epoch: ", i, "error : ", total_error/len(x_data))
```

```python

## 학습끝난 후 W와 b로 예측 하기
print("W : ", W)
print("b : ", b)
print("input 3 : ", tf.add(tf.multiply(3, W), b))
print("input 4 : ", tf.add(tf.multiply(4, W), b))
print("input 5 : ", tf.add(tf.multiply(5, W), b))
```
