```python
### 작성해주세요
### tf를 통해 변수를 선언하고 for문을 이용하여 1부터 10까지 더해지는 코드를 작성해주세요

import tensorflow as tf

## 변수 선언
x1 = tf.constant(0)

for i in range(1, 11):
    #x1 += i
    x1 += tf.constant(i)

print(x1)
## for 문


```
