```python
import pandas as pd
import sweetviz as sv
df = pd.read_csv("../dataset/pima-indians-diabetes.csv",header=None,
                 names = ['0', '1', '2','3', '4', '5','6', '7', 'TARGET'])

advert_report = sv.analyze(df)

#display the report
advert_report.show_html('./sweetviz_Advertising.html')

df1 = sv.compare([df.sample(300),"Train"], [df.sample(170),"Test"], "TARGET")
df1.show_html('./sweetviz_Compare_Target.html')
```
