```python

import tensorflow as tf
import pandas as pd
from sklearn.preprocessing import LabelEncoder
import numpy as np


```

```python
# 데이터 입력
df = pd.read_csv('../dataset/one_hot_encording_test.csv', names = ["length", "width", "height", "volume", "class_name"])
dataset=df.copy()

```

```python
Y_obj = dataset.pop("class_name")
X = dataset.copy()
print("data value : {}".format(X))
print("data Label: {}".format(Y_obj))

```

```python
# case 1)
# Label Encoder는 독립 변수가 아닌 종속 변수(라벨)에 대해 사용한다. 문자열이나 정수로된 라벨 값을  0  ~  K−1 까지의 정수로 변환.
e = LabelEncoder()
e.fit(Y_obj)
print("Label Class String : {}".format(e.classes_))
Y = e.transform(Y_obj)
print("Label Class Int: {}".format(Y))

Y_encoded = tf.keras.utils.to_categorical(Y)
print("case 1 One Hot Label Class : {}".format(Y_encoded))
print(np.argmax(Y_encoded, axis=1).reshape(-1,1))
print(Y_encoded.shape)

```

```python
## case 2)
one_hot_label=pd.get_dummies(Y_obj)
print("case 2 one_hot_label : ", one_hot_label)
print(one_hot_label.shape)
print(one_hot_label.values)
```
