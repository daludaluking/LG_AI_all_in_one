```python

import tensorflow as tf
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
import pandas as pd

from sklearn.preprocessing import LabelEncoder
df = pd.read_csv('sample_data/sonar.csv')

dataset = df.values
X = dataset[:,0:60].astype(float)
Y_obj = dataset[:,60]

e = LabelEncoder()
e.fit(Y_obj)
Y = e.transform(Y_obj)



```

```python

# 학습 셋과 테스트 셋의 구분
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, shuffle=True)
```

```python

## 모델 설계하기.
activation=tf.keras.activations.relu
input_Layer = tf.keras.layers.Input(shape=(60,))
x = tf.keras.layers.Dense(1000, activation=activation)(input_Layer)
x= tf.keras.layers.Dense(1000, activation=activation)(x)
x= tf.keras.layers.Dense(1000, activation=activation)(x)
x= tf.keras.layers.Dense(700, activation=activation)(x)
x= tf.keras.layers.Dense(700, activation=activation)(x)
x= tf.keras.layers.Dense(500, activation=activation)(x)
x= tf.keras.layers.Dense(500, activation=activation)(x)
x= tf.keras.layers.Dense(500, activation=activation)(x)
Out_Layer= tf.keras.layers.Dense(1, activation='sigmoid')(x)

model = tf.keras.Model(inputs=[input_Layer], outputs=[Out_Layer])
model.summary()
```

```python

# 모델 컴파일
loss=tf.keras.losses.binary_crossentropy
optimizer = tf.keras.optimizers.SGD(learning_rate=0.0009)
metrics=tf.keras.metrics.binary_accuracy
model.compile(loss=loss,
             optimizer=optimizer,
             metrics=[metrics])

# 모델 실행
model_hist=model.fit(X_train, Y_train, validation_split=0.2, epochs=1000, batch_size=100)
print(model_hist.history.keys())
```

```python

epochs = range(1, 1001)
original_val_loss = model_hist.history['val_loss']
original_loss = model_hist.history['loss']
plt.subplot(211)  ## 2x1 개의 그래프 중에 1번째
plt.plot(epochs, original_val_loss, 'b+', label='val_loss')
# ‘bo’는 파란색 점을 의미합니다
plt.plot(epochs, original_loss, 'ro', label='loss')
plt.xlabel('Epochs')
plt.ylabel('loss')
plt.legend()

original_val_loss = model_hist.history['val_binary_accuracy']
original_loss = model_hist.history['binary_accuracy']
plt.subplot(212)  ## 2x1 개의 그래프 중에 1번째
plt.plot(epochs, original_val_loss, 'b+', label='val_binary_accuracy')
# ‘bo’는 파란색 점을 의미합니다
plt.plot(epochs, original_loss, 'ro', label='binary_accuracy')
plt.xlabel('Epochs')
plt.ylabel('acc')
plt.legend()

# 테스트셋에 모델 적용
print("\n Test Accuracy: %.4f" % (model.evaluate(X_test, Y_test)[1]))

plt.show()
```
