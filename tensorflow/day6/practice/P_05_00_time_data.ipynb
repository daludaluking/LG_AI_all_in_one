```python


import tensorflow as tf
from sklearn.model_selection import train_test_split
import os
import pandas as pd

gpus = tf.config.experimental.list_physical_devices('GPU')
if gpus:
  try:
    # Currently, memory growth needs to be the same across GPUs
    for gpu in gpus:
      tf.config.experimental.set_memory_growth(gpu, True)
    logical_gpus = tf.config.experimental.list_logical_devices('GPU')
    print(len(gpus), "Physical GPUs,", len(logical_gpus), "Logical GPUs")
  except RuntimeError as e:
    # Memory growth must be set before GPUs have been initialized
    print(e)

data_dir = '../dataset'
fname = os.path.join(data_dir, 'data-02-stock_daily.csv')

```

```python
## 데이터 읽어오기.
df = pd.read_csv(fname)
dataset = df.values
ori_X = dataset[:,0:4]
ori_Y = dataset[:,4]

print(dataset.shape)
print(ori_X)
print(ori_Y)
```

```python
seq_length=8
X_train, X_test, Y_train, Y_test = train_test_split(ori_X,ori_Y, test_size=0.2, shuffle=False)
data_gen=tf.keras.preprocessing.sequence.TimeseriesGenerator(X_train, Y_train,
                               length=seq_length, sampling_rate=2,stride=8,
                               batch_size=1,
                               start_index=0,reverse=False) #reverse=False 순차적, TRUE는 역순:과거데이터를 먼저 넣는 모델을 만든다.
```
